```python
import pandas as pd
import numpy as np
```

```python
disease_df = pd.read_csv('Resources/TCM disease.csv')
disease_df.head()
```

```python
herbs_df = pd.read_csv('Resources/TCM herbs.csv')
herbs_df.head()
```

```python
herbs_df.columns
```

```python
disease_names = disease_df.get('Disease_name')
df_disease_columns = []
df_herb_columns = []

for d in disease_names:
    try:
        find_d = herbs_df[herbs_df['Indication'].str.contains(d, na=False, case=False)]
        herb_names = find_d['Herb_latin_name'].values
        if len(herb_names) > 0:
            df_disease_columns.append(d)
            df_herb_columns.append(set(herb_names))
    except:
        pass
```

```python
herb_disease_df = pd.DataFrame({
    'Disease': df_disease_columns,
    'Herb': df_herb_columns
})

herb_disease_df.head()
```

```python
explode_df = herb_disease_df.explode('Herb').reset_index(drop=True)
explode_df
```

```python
explode_df.to_csv('Resources/disease_herb.csv')
```

```python
top_diseases = explode_df['Disease'].value_counts()
top_diseases.to_csv('Resources/top_diseases.csv')
```
